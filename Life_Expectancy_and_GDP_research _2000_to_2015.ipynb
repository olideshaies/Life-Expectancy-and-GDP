##  Bar Charts To Compare Average

What is the average life expectancy in these nations?

```python
df_chile = df[df.Country == "Chile"]
df_china = df[df.Country == "China"]
df_germany = df[df.Country == "Germany"]
df_mexico = df[df.Country == "Mexico"]
df_usa = df[df.Country == "United Stated of Americ"]
df_zimbabwe = df[df.Country == "Zimbabwe"]

ax = sns.barplot(x="Country", y="LEABY", data=df)
plt.xticks(rotation=90)
```

```python
f, ax = plt.subplots(figsize=(15, 12)) 
ax = sns.barplot(x="Country", y="LEABY", hue="Year", data=df, palette=sns.dark_palette("muted purple", input="xkcd"))
plt.xticks(rotation=90)
ax.set(ylabel="Life expectancy at birth in years")
plt.show()
```

What is the average GDP in these nations?

```python
ax = sns.barplot(x="Country", y="GDP", data=df)
plt.xticks(rotation=90)
```

```python
f, ax = plt.subplots(figsize=(15, 12)) 
ax = sns.barplot(x="Country", y="GDP", hue="Year", data=df, palette=sns.color_palette("Paired"))
plt.xticks(rotation=90)
ylabel="GDP in Trillions of U.S. Dollars"
plt.ylabel(ylabel)
# ax.set(ylabel="Life expectancy at birth (years)")
plt.show()
plt.clf()

sns.barplot(x="Country", y="GDP", hue="Year", data=df_zimbabwe, palette=sns.color_palette("Paired"))
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
def trillions(x, pos):
    'The two args are the value and tick position'
    return '$%1.1fT' % (x*1e-12)

formatter = FuncFormatter(trillions)


f, ax = plt.subplots(figsize=(15, 6)) 
ax = sns.pointplot(x="Year", y="GDP", hue = "Country", data=df)
plt.xticks(rotation=70)
ax.yaxis.set_major_formatter(formatter)
ax.set(ylabel="GDP in Trillions of U.S. Dollars")
plt.show()
```

What is the distribution of that life expectancy?
